# Dealing with data type problems

```python
# Reminder of the DataFrame
cars.head()
```

En premier lieu on remplace DH dans la colonne Price par "," 

```python
# Replace "DH" with ","
cars['Price'] = cars['Price'].str.replace("DH",",")
# Print the header of the column
cars['Price'].head()
```

```python
# Split column into two
div = cars['Price'].str.split(",", expand = True)
div.head()
```

```python
# Assign correct columns to price and sale_type columns in cars
cars['price'] = div[0]
cars['sale_type'] = div[1]
# Replace the empty values in the "Sale_Type" column with "no info".
cars['sale_type'] = cars['sale_type'].replace('', 'no info')
# Delete "*" and "* *" from column sale_type
cars['sale_type'] = cars['sale_type'].str.strip("* *")
cars['sale_type'] = cars['sale_type'].str.strip("*")
# Print the header and confirm new column creation
cars.head()
```

```python
# Drop Price column
cars.drop('Price', axis = 1, inplace = True)
cars.head()
```

```python
# Print out dtypes again
cars.dtypes
```

```python
# Remove whitespace from the 'price' column
cars['price'] = cars['price'].str.strip()
# Convert the values of the 'price' column to float
cars['price'] = pd.to_numeric(cars['price'], errors='coerce')
print(cars.head())
```

```python
# Print dtypes again
cars.dtypes
```

```python
                                ##dealing with 'Puissance fiscale' column

    
# Remove "cv" from 'Puissance fiscale' before conversion to float
cars['Puissance fiscale'] = cars['Puissance fiscale'].str.strip("cv")


# Remove whitespace from the 'Puissance fiscale' column
cars['Puissance fiscale'] = cars['Puissance fiscale'].str.strip()

#replacing missing values and dashes in the 'Puissance fiscale' column with 0 any empty strings with 0.
cars['Puissance fiscale'] = cars['Puissance fiscale'].fillna('0').replace('-', '0')
cars['Puissance fiscale'] = cars['Puissance fiscale'].replace('', '0')

# Convert Puissance fiscale to float
cars['Puissance fiscale'] = cars['Puissance fiscale'].astype('float')

# Print header to make sure change was done
cars['Puissance fiscale'].head()
```

```python
                              ##dealing with 'Kilométrage' column


# Remove "km" from kilométrage before conversion to float
cars['Kilométrage'] = cars['Kilométrage'].str.strip("km")

# Remove whitespace from the 'Kilométrage' column
cars['Kilométrage'] = cars['Kilométrage'].str.strip()

# Convert Kilométrage to float
cars['Kilométrage'] = pd.to_numeric(cars['Kilométrage'], errors='coerce')

# Print header to make sure change was done
cars['Kilométrage'].head()
```

```python
                            ##dealing with 'Cylindrée' column

    
# Remove "cm³" from Cylindrée before conversion to float
cars['Cylindrée'] = cars['Cylindrée'].str.strip("cm³")

#replacing missing values and dashes in the 'Cylindrée' column with 0
cars['Cylindrée'] = cars['Cylindrée'].replace('', '0')

# Convert Cylindrée to float
cars['Cylindrée'] = cars['Cylindrée'].astype('float') 

# Print header to make sure change was done
cars['Cylindrée'].head()
```

```python
                          ##dealing with 'Volume du réservoir' column

# Remove "litre" before conversion to float
cars['Volume du réservoir'] = cars['Volume du réservoir'].str.strip("litre")

#replacing missing values and dashes with 0 any empty strings with 0.
cars['Volume du réservoir'] = cars['Volume du réservoir'].fillna('0').replace('-', '0')
cars['Volume du réservoir'] = cars['Volume du réservoir'].str.strip()
cars['Volume du réservoir'] = cars['Volume du réservoir'].replace('', '0')

# Convert to float
cars['Volume du réservoir'] = cars['Volume du réservoir'].astype('float')

# Print header to make sure change was done
cars['Volume du réservoir'].head()
```

```python
                          ##dealing with 'Vitesse maxi.' column
    
# Remove "km/h" before conversion to float
cars['Vitesse maxi.'] = cars['Vitesse maxi.'].str.strip("km/h")

#replacing missing values and dashes with 0 any empty strings with 0.
cars['Vitesse maxi.'] = cars['Vitesse maxi.'].str.strip()
cars['Vitesse maxi.'] = cars['Vitesse maxi.'].replace('', '0')

#replacing all occurrences of '-' in the 'Vitesse maxi.' column with a NaN 
cars['Vitesse maxi.'] = cars['Vitesse maxi.'].replace('-', float('nan'), regex=True)

# Convert to float
cars['Vitesse maxi.'] = cars['Vitesse maxi.'].astype('float')

# Print header to make sure change was done
cars['Vitesse maxi.'].head()
```

```python
cars.dtypes
```

```python
# Remove "l/100 km" before conversion to float
cars['Conso. ville'] = cars['Conso. ville'].str.strip("l/100 km")
cars['Conso. route'] = cars['Conso. route'].str.strip("l/100 km")
```

```python
cars['Conso. ville'].head()
cars['Conso. route'].head()
```

```python

#convert the 'Conso. ville' and 'Conso. route' columns from strings to floats, and replaces any non-numeric values with NaN.

cols_to_clean = ['Conso. ville', 'Conso. route']
for col in cols_to_clean:
    cars[col] = cars[col].apply(lambda x: float(str(x).strip().replace(',', '.')) if str(x).strip().replace(',', '.').replace('.', '').isdigit() else float('nan'))
```

```python
cars.dtypes
```

```python
                               ## dealing with 'Architecture' column
    

cars['Architecture'] = cars['Architecture'].str.strip("cylindres en ligne")
cars['Architecture'] = cars['Architecture'].str.strip("cylindres")
cars['Architecture'] = cars['Architecture'].str.strip("cylindres à plat")
cars['Architecture'] = cars['Architecture'].str.strip("cylindres en V")

# Convert to float
cars['Architecture'] = cars['Architecture'].astype('float')
```

## Partie 3 : Construction d'un modèle de prédiction
1- Construire un modèle de prédiction pour prédire le prix des voitures en utilisant les caractéristiques restantes du prétraitement.

2- Appliquer la sélection des caractéristiques à votre ensemble de données nettoyé et reconstruire votre modèle.

3- Evaluer et comparer les deux modèles

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import pandas as pd

# Load data
cars = pd.read_csv('wandaloo_cars.csv')

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(cars.drop('price', axis=1), cars['price'], test_size=0.2, random_state=42)

# Convert categorical variables to dummy variables
X_train = pd.get_dummies(X_train, drop_first=True)
X_test = pd.get_dummies(X_test, drop_first=True)

# Fit linear regression model
lin_reg = LinearRegression()
lin_reg.fit(X_train, y_train)

# Predict response variable for test set
y_pred = lin_reg.predict(X_test)

# Evaluate model performance
from sklearn.metrics import mean_squared_error, r2_score

# Mean squared error (MSE)
mse = mean_squared_error(y_test, y_pred)
print("Mean squared error (MSE): {:.2f}".format(mse))

# R-squared score (coefficient of determination)
r2 = r2_score(y_test, y_pred)
print("R-squared score: {:.2f}".format(r2))
```
